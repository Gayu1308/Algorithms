```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_csv('/content/Book1.csv')
```

```python
db = DBSCAN(eps=50, min_samples=2)
labels = db.fit_predict(data)
```

```python
df = pd.DataFrame(data, columns=['Readings', 'Month'])
df['Label'] = labels
print("DBSCAN Output:\n")
print(df)
plt.figure(figsize=(8,5))
colors = ['green' if label != -1 else 'red' for label in labels]
plt.scatter(data['Readings'], data['Month'], c=colors, s=100, edgecolors='black')
plt.title('DBSCAN Anomaly Detection')
plt.xlabel('Readings(Units)')
plt.ylabel('Month')
plt.grid(True)
plt.show()
```
