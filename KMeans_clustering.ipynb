```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/content/credit card_new.csv')
```

```python
data.info()
```

```python
X = data[['Time','Amount','V28']]
```

```python
# Drop rows with missing values
X_cleaned = X.dropna()

kmeans = KMeans(n_clusters=5, random_state=42, n_init='auto')
labels= kmeans.fit_predict(X_cleaned)
```
